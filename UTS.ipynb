```python
df = pd.read_csv('/content/Ganjil.csv')

def text_cleaning(text):
    text = str(text).lower()
    text = re.sub(r'http\S+|www\S+', '', text)  # hapus URL
    text = re.sub(r'[^a-zA-Z\s]', '', text)    # hapus angka & simbol
    text = re.sub(r'\s+', ' ', text).strip()   # hapus spasi berlebih
    return text

df['clean_text'] = df.iloc[:,0].apply(text_cleaning)
```

```python
nltk.download('punkt_tab')
df['tokens'] = df['clean_text'].apply(word_tokenize)
```

```python
normal_dict = {
    "gk": "tidak",
    "ga": "tidak",
    "nggak": "tidak",
    "aja": "saja",
    "yg": "yang",
    "dr": "dari",
    "tdk": "tidak"
}

def normalize(tokens):
    return [normal_dict[word] if word in normal_dict else word for word in tokens]

df['normalized'] = df['tokens'].apply(normalize)
```

```python
stop_words = set(stopwords.words('indonesian'))

def remove_stopwords(tokens):
    return [word for word in tokens if word not in stop_words]

df['no_stopwords'] = df['normalized'].apply(remove_stopwords)
```

```python
factory = StemmerFactory()
stemmer = factory.create_stemmer()

def stemming(tokens):
    return [stemmer.stem(word) for word in tokens]

df['stemmed'] = df['no_stopwords'].apply(stemming)
```

```python
lemmatizer = WordNetLemmatizer()

def lemmatization(tokens):
    return [lemmatizer.lemmatize(word) for word in tokens]

df['lemmatized'] = df['stemmed'].apply(lemmatization)
```
